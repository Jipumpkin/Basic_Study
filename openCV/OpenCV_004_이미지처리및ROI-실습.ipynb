# 이미지 처리 방법
## 픽셀기반 처리
- 이미지는 픽셀이라는 데이터 유닛의 집합
- R, G, B 세가지 채널을 가지고 있음(png 데이터는 알파까지 4채널)
- c * h * w의 매트릭스

## 블록기반 처리
- 블록 : 픽셀을 그룹화 한 것

## 주파수 기반 처리
- 이미지에서의 주파수 : 이미지 픽셀값의 변화
- 고주파 : 픽셀값의 변화가 큼
- 저주파 : 픽셀값의 변화가 작음


```python
# 실습 1 - 이미지 픽셀값 출력
import cv2

image = cv2.imread('images/01_lenna.png')
#print(image)

px = image[200,100]
print(px)
```

```python
# 실습 2 - 픽셀값 변경

import cv2
import matplotlib.pyplot as plt

cv2.startWindowThread()
image = cv2.imread('images/01_lenna.png')
#image[200,100] = [0, 0, 0]
image[200, :, :] = [0]
b, g, r = cv2.split(image)
new_img = cv2.merge([r, g, b])

plt.imshow(new_img)
plt.show()
```

```python
# 실습 3 - RGB 채널

import cv2
import matplotlib.pyplot as plt

# split은 채널을 분리
cv2.startWindowThread()
image = cv2.imread('images/01_lenna.png')
b, g, r = cv2.split(image)

# merge는 채널을 합침
new_img = cv2.merge([r, g, b])
plt.imshow(new_img)
plt.show()
```

## 블록 기반 처리 ROI(Region of Image)

```python
import cv2

cv2.startWindowThread()
image = cv2.imread('images/01_lenna.png')

subimg = image[35:300,150:350]
b, g, r = cv2.split(subimg)
new_img = cv2.merge([r, g, b])
plt.imshow(new_img)
plt.show()
```

## 도형 그리기

- <font color=red>cv2.circle : 원</font>
- <font color=red>cv2.rectangle : 사각형</font>
- <font color=red>cv2.elipse : 타원</font>
- <font color=red>cv2.line : 선</font>
  - 사용방법 매개변수4개( 배경이미지, 좌표, 색상, 선의 두께)
- <font color=red>cv2.putText(배경이미지, 출력내용, 출력시작좌표, 폰트, 크기, 색상, 굵기)</font>

```python
# 실습 - 선 그리기

import cv2

cv2.startWindowThread()
img = cv2.imread('images/01_lenna.png')

cv2.line(img, (50, 150), (50, 200), (0, 0, 255), thickness=5, lineType=None, shift=None)
b, g, r = cv2.split(img)
img = cv2.merge([r, g, b])

plt.imshow(img)
plt.show()
```

```python
# 실습 - 사각형 그리기

import cv2

img = cv2.imread('images/01_lenna.png')

img = cv2.rectangle(img, (180,250),(350,430),(0,0,255),3)
font = cv2.FONT_HERSHEY_SIMPLEX
img = cv2.putText(img,'Hello Lena', (180,200), font, 1, (255,0,0), 4)
b, g, r = cv2.split(img)
img = cv2.merge([r, g, b])
plt.imshow(img)
plt.show()
```

```python
# 실습 - 원 그리기

import cv2

img = cv2.imread('images/01_lenna.png')

cv2.circle(img, (180,250), 100, (0,0,255), 3)
font = cv2.FONT_HERSHEY_SIMPLEX
img = cv2.putText(img,'Hello Lena', (100,250), font, 1, (255,0,0), 4)
b, g, r = cv2.split(img)
img = cv2.merge([r, g, b])
plt.imshow(img)
plt.show()
```
