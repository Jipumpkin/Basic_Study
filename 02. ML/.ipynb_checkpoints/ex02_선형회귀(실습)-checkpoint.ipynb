##### 결측치 처리하기
- 결측치가 있는 데이터를 삭제하거나 결측치를 대체값으로 채우는 선택을 할 수 있다.
- 데이터 양이 많은 경우 결측치를 삭제해도 무방하나 데이터 양이 적다면 대체값을 채우는게 좋은 선택이 될수있다.
  단 대체값의 경우 올바른 값이 아니면 분석을 방해할 가능성이 있다

```python
# 결측치가 있는 특성 확인
house_data.isnull().sum().sort_values(ascending=False)/len(house_data)*100 # 컬럼별 결측치의 비율
# 결측치 비율이 높은 컬럼들은 활용할때 신중히 고민해야한다 -> 대체값 채울떄도 정교하게 진행해야함
# 결측치 비율이 상대적으로 낮은 CouncliArea와 Car를 전처리해보자
```

```python
# 기술통계를 확인하여 결측치 채우기
house_data['Car'].describe()
```

```python
# 결측치를 채우는 함수 -> fillna
house_data['Car'] = house_data['Car'].fillna(0)
```

```python
# 결측치 비율 재확인
house_data.isnull().sum().sort_values(ascending=False)/len(house_data)*100 # 컬럼별 결측치의 비율
```

```python
# Suburb(지역이름)을 통해서 CouncilArea(관할구역) 결측치를 채워보자
house_pivot = house_data.pivot_table(index = ['Suburb','CouncilArea'], # 행에 표현될 컬럼 선택
                                     values = 'Rooms', # 결측치가 없는 컬럼 선택(다른것도 가능)
                                     aggfunc = 'count') # 갯수를 세주는 집계함수 설정
```

```python
# 하나의 지역이 2개 이상의 관할구역에 배정되어 있는 경우가 있음
# 개인선택 : 빈도가 높은 관할구역으로 결측치가 채워지도록 함
house_pivot.head(50)
```

```python
# 최빈값으로 구성된 테이블 생성
Council_mode_table = house_data.groupby('Suburb')['CouncilArea'].agg(pd.Series.mode)
Council_mode_table
```

```python
# apply 함수를 이용해서 각 행마다 결측치를 처리
def fill_council(row):
    if pd.isnull(row['CouncilArea']) : # 결측치인 경우
        return Council_mode_table[row['Suburb']]
    else :
        return row['CouncilArea']
```

```python
house_data['CouncilArea'] = house_data.apply(fill_council, axis=1)
```

```python
# 결측치 비율 재확인
house_data.isnull().sum().sort_values(ascending=False)/len(house_data)*100 # 컬럼별 결측치의 비율
```

```python
# 각 컬럼과 Price 사이의 상관계수 확인
house_data.corr(numeric_only=True)['Price'].sort_values()
```

##### 학습을 위한 전처리
- 특성선택
- 글자 -> 숫자 변경(인코딩)
- 지도학습을 위한 문제, 답 분리
- 훈련,평가 데이터 분리

```python
# 학습에 사용할 입력특성 선택
feature_names = ['Rooms','Bedroom2','Bathroom','Lattitude','Longtitude','Type_encoding'] # 코드블럭 나중에 실행해야함
```

```python
# 인코딩 작업
# 인코딩 방법 2가지 => 1.원핫인코딩 2.레이블인코딩
# 원핫인코딩을 쓰는 경우 : 해당컬럼이 순서하는 의미가 없는 경우
# 레이블인코딩을 쓰는 경우 : 해당컬럼에 순서가 부여된 경우(ex 점수등급 A,B,C,D)
```

```python
# type은 평균가격이 순서가 있기 때문에 레이블인코딩을 써볼 수 있겠다.
house_data['Type_encoding'] = house_data['Type'].map({'h':0,'t':1,'u':2})
```

```python
house_data['CouncilArea'] = house_data['CouncilArea'].astype('str')
# 문자열로 변환
```

```python
# '[]' 가 들어있는 데이터의 index 추출
drop_index = house_data[house_data['CouncilArea']=='[]'].index
drop_index
```

```python
house_data.drop(drop_index, inplace=True)
```

```python
# CouncilArea는 원핫인코딩을 이용해서 동등한 크기를 지니는 숫자로 인코딩
council_one_hot = pd.get_dummies(house_data['CouncilArea'], dtype="int64", prefix="council") # prefix 컬럼 이름 중복방지를 위한 출처(접두어)표시
council_one_hot
```

```python
# 선택한 특성 및 인코딩한 특성을 병합
df1 = house_data[feature_names]
df1.head()
```

```python
# 문제(입력특성)
X = pd.concat([df1, council_one_hot], axis=1)
# 정답
y = house_data['Price']
```

```python
X.shape, y.shape
```

```python
# 훈련용, 평가용으로 분리
# 학습된 모델의 신뢰도를 확보하기 위해 평가용 데이터를 분리
# 보통 7:3의 비율로 분할
# 가능하다면 평가용 데이터가 여러 셋트로 확보되면 좋다

from sklearn.model_selection import train_test_split as tts

# train_test_split(문제, 정답, 분리비율, 고정키)

X_train, X_test, y_train, y_test = tts(X, y, test_size= 0.3, random_state= 15)
```

```python
# 크기확인
X_train.shape
```

```python
# 데이터 크기확인
print(f"훈련용 문제 : {X_train.shape}\n"
      f"훈련용 답 : {y_train.shape}\n"
      f"테스트용 문제 : {X_test.shape}\n"
      f"테스트용 답 : {y_test.shape}")
```

#### 4. 모델선택 및 학습

```python
# 모델생성
from sklearn.linear_model import LinearRegression as LR
house_model_linear = LR()
```

```python
house_model_linear.fit(X_train, y_train)
```

```python
# 가중치, 절편 확인하기
# y = wx + b
print(f"가중치 : {house_model_linear.coef_}")
print(f"절편 : {house_model_linear.intercept_}")
```

#### 5. 모델평가

```python
# 모델 평가
# score 로 점수 내주기
print(f"Linear Score : {house_model_linear.score(X_test, y_test)}")
```

```python
# 평균제곱오차 metrics : 평가지표 모음집
from sklearn.metrics import mean_squared_error as MSE
```

```python
# 평가지표
linear_pre_test = house_model_linear.predict(X_test)
print(f"mse : {MSE(y_test, linear_pre_test)}") # 실제값, 예측값 차이
print(f"rmse : {MSE(y_test, linear_pre_test) ** 0.5}")

# 제곱값ㅇ느 실제 오차에 제곱이 되므로 실제를 알아보기 힘듦
# 실제 단위 달러의 제곱근을 씌우기 ( **0.5)
# 40만 달러 차이
```
