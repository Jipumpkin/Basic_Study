```python
# def 키워드를 사용해서 함수를 정의(define)해서 사용
# 매개변수가 없으면 생력 가능
# 리턴값도 없으면 생략 가능
```

```python
# 함수 사용 방법ㅂ
# 1. 형식에 맞게, 사용법에 맞게 함수를 정의
# 2. 사용을 위해 정의된 함수 이름을 호출
# 3. 매개변수인자가 있다면 함수 안에 넣고 리턴값을 받아내면 끝
```

```python
def number_sum(num1, num2): # 매개변수인자
    result = num1 + num2 # 들어온 매개변수를 통해서 실행문장 구성
    return result # 실행결과 리턴
```

```python
number_sum(2,10)
```

```python
def divisor(num) :
    for i in range(1,num):
        if num%i == 0:
            print(i, end= " ")
```

```python
divisor(100)
```

```python
def greeting(name):
    """ 이름이 입력되면 인사를 하는기능을 가지 ㄴ함수
    name type : str
    return type : str
    """
    return name + ', 반갑다.'
```

### 가변매개변수(*)
- 함수를 호출할때 몇개의 인수가 전달될지 알 수 없다면, 사용자가 직접 매개변수의 개수를 정할 수 있도록 선언

```python
# sum(1,2,3) -- > 오류
sum([1,2,3]) # 리스트 요소 한개만 들어왔을때
```

```python
def add(*srgs):
    return sum(srgs)
```

```python
add(1,2,3,4,5)
```

### 딕셔너리 형태의 가변 매개변수(**)
- 딕셔너리 형태로 함수 내부를 처리하고 싶을때

```python
def print_map(**kwargs):
    print(kwargs)
    for k, v in kwargs.items():
        print(k,":",v)
```

```python
print_map(하나 = 1, 둘 = 2)
```

### lambda 함수
- lambda는 함수를 생성할 때 사용하는 예약어로 def와 동일한 역할을 함
- 람다라고 읽고 익명의 함수라고도 함
- 간단하게 정의할 필요가 있는 함수를 쓸때 사용
- return 명령어가 필요 없다 -> 표현식이 마무리되면 결과값을 돌려주는 함수

```python
# lambda 함수는 익명의 함수
def f(x):
    return x * 2
f(3)
```

```python
g = lambda x : x*2
g(3)
```

```python
# 매개변수가 여러개일때
h = lambda a,b,c : (a+b)*2
h(1,2,3)
```

```python
# 당첨번호를 랜덤으로 추출하는 기능
import random
lotto = []
while True:
    num = random.randint(1,45)
    if num not in lotto:
        lotto.append(num)
    if len(lotto) == 6:
        break;

lotto
```

```python
# 로또번호를 띄어쓰기로 입력
inputStr = input("내 로또번호를 공백을 기준으로 6개 입력").split(" ")
mine = []
for n in inputStr:
    mine.append(int(n))
```

```python
# 내 번호가 lotto 당첨번호 리스트에 포함되어있는지 체크
same = 0
for myN in mine:
    if myN in lotto:
        same += 1
```

```python
# same에 따라 인덱싱에서 등수 표기
score = [0,6,5,3,2,1]
if same == 0:
    print('탈락')
else :
    print(f'{same}개 맞음, {score[same]}등 당첨!')
```

```python
import numpy as np
def make_lotto(count):
    for i in range(count):
        lotto_num = [] # 로또 번호가 담길 리스트형 변수

        for j in range(6): # 6번반복
            lotto_num = np.random.choice(range(1,46), 6, replace=False)
        lotto_num.sort() # 작은 숫자가 먼저 나오도록 정렬
        print(f"{i+1}. 로또번호 : {lotto_num}")
```

```python
count = int(input("로또 번호를 몇 개 생성할까요? >> "))
make_lotto(count)
```
