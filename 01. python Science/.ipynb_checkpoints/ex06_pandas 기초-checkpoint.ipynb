#### 연습문제
- 육균신체측정정보 데이터를 이용해서 BMI지수 계산하기

```python
육신정 = pd.read_csv("./육군신체측정정보.csv", encoding = 'utf-8', index_col = '순번')
육신정
```

```python
# 데이터 정보 확인
육신정.info()
```

```python
육신정['신장 센티미터'] = 육신정['신장 센티미터'].str.replace(' cm', '').astype(float)
육신정
```

```python
육신정['몸무게 킬로그램'] = 육신정['몸무게 킬로그램'].str.replace(' kg', '').astype(float)
육신정.info()
```

```python
BMI = 육신정['몸무게 킬로그램'] / (육신정['신장 센티미터']*0.01)**2
육신정['BMI지수'] = BMI.round(2)
육신정
```
