### 변수
- 사전적의미 : '변할 수 있는 수'
- 프로그래밍적의미 : 데이터를 담을 수 있는 공간

```python
num = 3
print(num)
print(type(num))
```

```python
# 변수명을 만들 때 규칙
# 1. 영문자, 숫자 , 언더바(_)를 사용
# 2. 영문자는 대문자와 소문자를 다르게 인식(구분)
num = 5
Num = 10
print(num)
print(Num)
```

```python
# 3. 변수명은 숫자로 시작하면 안된다.
1num = 10
print(1num)
```

```python
# 4. 키워드(예약어) 사용 불가능
False = 10
# 예약어인지 아닌지 구분법
# 변수명이 초록색 -> 예약어
# -> 변수로 사용가능한 문자는 검은색
```

```python
# 파이썬의 특별한 변수 대입 방법
a1, b1 = 15, 23
print(a1, b1)
a2 = b2 = 7
print(a2, b2)
```

### 문자열

```python
# 파이썬에서는 문자열을 사용하는 기호로 작음따옴표와 큰따옴표 모두 사용 가능
s = 'si'
b = "bal"
print(s)
print(b)

um = "si\nsisisisisisisisi"
print(um)
```

### 문자열의 인덱싱과 슬라이싱
- 인덱스 : 색인, 목차 -> 순서가 있다는 뜻
- 인덱싱 : 무엇인가를 가리킨다.
- 슬라이싱 : 범위를 지정해서 어디서부터 어디까지 잘라낸다

```python
# 인덱싱 -> 기호 : 대괄호
s = "My Name is bingsin"
print(s)
```

```python
# 슬라이싱
# 'my', 'name' 문자열 가져오기
s = "my name is sibal"
print(s[0:2]) # 슬라이싱 -> s[a:b] -> a이상, b미만인 인덱스 값까지 가져온다.
print(s[3:7])
```

```python
# my name 가져오기
print(s[:7])
print(s[8:])
```

### 문자열 특정 부분 초기화(formatting)

```python
# f문자열 포매팅
month = 8
day = 5
s = f'오늘은 {month}월 {day}일 입니다'
print(s)
```
