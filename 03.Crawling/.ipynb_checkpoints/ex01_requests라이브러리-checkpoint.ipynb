### 학습목표
- requests 라이브러리 개념에 대해 알고 requests 라이브러리를 활용하여 페이지의 정보를 요청, 응답받기
- requests 라이브러리
  - 서버에게 데이터를 요청하여 응답 받기위한 라이브러리 (브라우저의 역할을 대신함)
  - 페이지 정보를 요청하고 응답받으면 역할이 끝

```python
# 라이브러리 불러오기
import requests as req
```

```python
# 네이버 페이지 정보 받아오기
url = 'https://www.naver.com/'
# 네이버 url을 통하여 페이지 정보 받아오기
res = req.get(url)

# Response [200] : 성공적으로 정보를 입력받았다는 의미
# Response [400] : 클라이언트 pc에 요청문제가 있을 경우의 출력 결과
# Response [500] : 서버(응답) 에 문제가 있을때 출력 결과
# 오류를 해결할때 도움이 된다.
```

```python
# 쿠팡페이지 정보 받아오기

# req.get('https://www.coupang.com/')

```

```python
# 응답받은 페이지 정보 확인
res.text
# 우리가 웹페이지에서 데이터를 수집하기위해 윕페이지를 구성하는 html언어를 수집
# 현재 우리가 ㅇ응답받은 데이터는 문자열 형태의 데이터이다
# 컴퓨터가 알아먹을수 있는 html 코드로 변경해주는 작업이 필요하다.
```

### BeautifulSuop 라이브러리
- 컴퓨터가 알아먹을 수 있는 언어로 변경해주는 라이브러리
- 응답받은 문자열 형태의 데이터를 html형태로 변경해주는 라이브러리

```python
from bs4 import BeautifulSoup as bs
                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                          
```

```python
# bs(파싱할 데이터,파싱방법) -> beautifulsoup 객체화
soup = bs(res.text, 'lxml')
```

```python
# '날씨' 를 검색한 페이지의 정보를 받아와 bs객체로
url2 = "https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query=%EB%82%A0%EC%94%A8"

res2 = req.get(url2)

soup2 = bs(res2.text, 'lxml')
```

```python
# '블로그' 라는 콘텐츠를 추출 -> 태그정보를 통해 데이터를 추출
# 페이지정보를 가진 변수.select('태그정보')
soup2.select('a')
# 페이지에 있는 모든 a 태그를 가져온것 -> 리스트에 담겨서 출력
# 내가 원하는 특정 요소만을 추출하기 위해서 구분자를 활용
# class(.), id(#) 선택자를 사용
```

```python
# 클래스가 tab인 a태그만을 추출
# select('태그명 기호 구분자이름')
data = soup2.select('a.tab')
data
# '블로그'만 추출하고 싶다면 -> 인덱싱
```

```python
# '블로그'를 포함한 요소만 추출
# 요소에서 콘텐츠만을 추출하는 방법
# 요소.text
data[0].text

# .text 속성은 요소에만 적용이 가능하다.
```

```python
# 전체 탭 출력
for i in range(len(data)):
    print(data[i].text)
```

```python
# 온도정보 가져오기
soup2.select('strong')
```

```python
# 구분자를 통해 다른 태그들과 구분하려 했으나 strong태그는 구분자를 갖지 않음
# 부모태그를 활용하여 다른 태그들과 구분해주는 작업이 필요하다.
# 주의할점은 반드시 최종목적지까지 정확하게 적어주는 것이 필요

soup2.select('div.temperature_text')
```

### 실습
- 네이버페ㅣ지에서 '가을'검색한 페이지 정보를 받아와
- 뉴스 제목을 수집

```python
Surl = "https://search.naver.com/search.naver?sm=tab_hty.top&where=nexearch&ssc=tab.nx.all&query=%EA%B0%80%EC%9D%84&oquery=%EB%82%A0%EC%94%A8&tqi=irEFewqVN8VsscVJNShssssstHh-216848"
gaul = req.get(Surl)

soup3 = bs(gaul.text, 'lxml')

Sdata = soup3.select('a.news_tit')

for i in range(len(Sdata)):
    print(Sdata[i].text)
```

```python
# 내가 가져오고싶은 데이터가 첫번째 데이터라면?
soup3.select_one('a.news_tit')
# select 와 select_one 의 차이점은?
# select: 태그정보를 가지는 모든 요소를 추출(리스트에 담아서)
# select_one: 태그정보를 가지는 가장 첫번째 요소만 추출
```

```python
soup3.select_one('a.news_tit').text
```
