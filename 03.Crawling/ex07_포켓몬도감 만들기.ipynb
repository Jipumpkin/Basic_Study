### 이미지 데이터 수집
- 이미지 데이터를 수집하여 폴더에 저장
- 수집한 이미지를 저장할 공간 (폴더) 생성하는 코드도 자동화

```python
# 라이브러리 불러오기

# ChromeDrierManager, Service 라이브러리를 사용하여
# Chromedriver(컴퓨터용 크롬창) 를 관리

from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

from selenium import webdriver as wb                 # 컴퓨터용 브라우저
from selenium.webdriver.common.keys import Keys      # 컴퓨터용 키보드
from selenium.webdriver.common.by import By          # 기준 -> 선택자 지정 방식

from bs4 import BeautifulSoup as bs # 태그 이름 보기위해 뷰티풀숲 사용


# 폴더 생성 라이브러리 -> 파일시스템 라이브러리
import os
# 이미지처리 라이브러리 -> 수집한 이미지 경로 (src) -> 이미지파일로 변경
from urllib.request import urlretrieve
import time
import pandas as pd
```

```python
# os 라이브러리를 활용하여 폴더 생성
# os.mkdir("경로/폴더명")


# 이미 존재하는 폴더 이름이라면 오류 발생
# 이미 있는 폴더이름이면 있다고 출력

if not os.path.isdir("C:/Users/USER/Desktop/포켓몬도감"):
    print("폴더생성")
    os.mkdir("C:/Users/USER/Desktop/포켓몬도감")
else :
    print("이미 존재하는 폴더 이름임")
```

```python
# 브라우저 열기

s = Service(ChromeDriverManager().install())

driver = wb.Chrome(service = s)
driver.get("https://www.pokemonkorea.co.kr/pokedex#pokedex_1")
```

```python
driver.find_elements(By.CLASS_NAME, 'img-fluid')
```

```python
# 이미지태그 요소 확인 -> bs 객체화 진행
# bs(페이지 정보, 파싱 방법)
soup = bs(driver.page_source, 'lxml')
```

```python
soup.select('img.img-fluid')[0]
```

```python
img = soup.select('img.img-fluid')
```

```python
# 속성값을 추출하는 방법
# 요소 ['src']
img[0]['src']
```

- 드라이버를 통해 요소 가져와서 이미지 수집 (속성추출)

```python
img_final = driver.find_elements(By.CLASS_NAME, 'img-fluid')
img_final[0].get_attribute('src')
```

```python
# 순수한 이미지 주소만을 담은 리스트 생성
img_list = []

for i in img_final:
    img_list.append(i.get_attribute('src'))
```

```python
img_list
```

```python
# 이미지 주소를 활용하여 이미지 파일로 저장
# urlretrieve(이미지 주소, 저장할 위치/파일명.확장자 )

# 첫번째 이미지만 이미지파일로 저장
urlretrieve(img_list[0], 'C:/Users/USER/Desktop/포켓몬도감/포켓몬.jpg')
```

```python
# 여러마리 포켓몬 저장
for i in range(len(img_list)):
    urlretrieve(img_list[i], f'C:/Users/USER/Desktop/포켓몬도감/포켓몬{i}.jpg')
```

- 더 많은 이미지 정보를 가져오기 위해 스크롤 진행

```python
# end키를 활용하여 스크롤 내리기
# 화면 전테를 담당하는 body태그에게 end 키보드를 전송
body = driver.find_element(By.CSS_SELECTOR, 'body')

for i in range(20):
    body.send_keys(Keys.END)
    time.sleep(1)
```

- 코드 합치기

```python
# 브라우저 열기

s = Service(ChromeDriverManager().install())

driver = wb.Chrome(service = s)
driver.get("https://www.pokemonkorea.co.kr/pokedex#pokedex_1")

# 스크롤 내리기
body = driver.find_element(By.CSS_SELECTOR, 'body')
for i in range(20):
    body.send_keys(Keys.END)
    time.sleep(1)

# 이미지 태그 가져오기
img_final = driver.find_elements(By.CLASS_NAME, 'img-fluid')

# 이미지 주소 리스트
img_list = []

# 이미지 주소를 리스트에 누적
for i in img_final:
    img_list.append(i.get_attribute('src'))

# 주소를 이미지 파일로 저장
for i in range(len(img_list)):
    urlretrieve(img_list[i], f'C:/Users/USER/Desktop/포켓몬도감/포켓몬{i}.jpg')
```
