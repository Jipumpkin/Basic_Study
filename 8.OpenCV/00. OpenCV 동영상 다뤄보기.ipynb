```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("./data/chip.jpg")

cv2.imshow("chip", img)
print(cv2.waitKey(5000)) # 언제까지 윈도우 창이 켜져있게 할건지 (단위: ms, 1s = 1000ms)
cv2.destroyAllWindows() # 창을 닫는 기능

# img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# plt.imshow(img)
# plt.show()
```

```python

# 동영상 불러오기
cap = cv2.VideoCapture(0) # VideoCapture 객체
# Video는 이미지들의 배열

cnt = 0 # 이미지 저장에 사용될 파일명 번호

while True :
    ch, frame = cap.read() # ch : 프레임을 잘 읽어왔는지 여부, frame : 읽어온 이미지(프레임)

    if not ch :
        print("동영상 재생 끝 or 읽기 실패")
        break
    cv2.imshow("video" ,frame)
    # 'q'키가 입력되면 동영상 재생 종료
    key = cv2.waitKey(30)
    if key == ord('q'):
        print("사용자에 의해서 종료됨")
        break

    # 'w'키가 입력되면 현제 프레임을 캡처하는 기능
    if key == ord('w'):
        cnt += 1
        print(f"./data/human_img{cnt}.png 사진 저장")
        cv2.imwrite(
            f"./data/human_img{cnt}.png", # 경로 및 파일명
            frame,
            params = [cv2.IMWRITE_PNG_COMPRESSION, 0] # 이미지 저장 옵션
        )

cap.release() # 자원 반납
cv2.destroyAllWindows()
```

```python
plt.rcParams['font.family'] = "Malgun Gothic" # plt 한글 설정

plt.suptitle("우리는 감자도리다") # 메인 타이틀

for i in range(1,5):
    selca = cv2.imread("./data/team.png")
    selca = cv2.cvtColor(selca, cv2.COLOR_BGR2RGB)

    # 서브플롯 설정
    # subplot(행, 열, 인덱스(공간))
    ax = plt.subplot(2,2,i)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.imshow(selca)

```
