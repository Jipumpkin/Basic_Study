```python
# transformers 안에있는 pipeline 기능
from transformers import pipeline

model = pipeline(task='text-classification', model="hun3359/klue-bert-base-sentiment")

text = "플라스크 수업은 fuuny"

model(text)
```

```python
# 플라스크와 모델 연동하기
# 라이브러리 불러오기
from flask import Flask, render_template, request
from transformers import pipeline
import json

# 모델불러오기

# transformers 안에있는 pipeline 기능
model = pipeline(task='text-classification', model="hun3359/klue-bert-base-sentiment")

# 텍스트를 입력받아 감정을 분석하여 예측값을 돌려주는 사용자 정의함수

def predict_emtion(text):
    result = model(text)
    return result

# 플라스크 웹 초기화

app = Flask(__name__)

@app.route('/')
def home():
    return render_template('predict.html')

# 예측한 결과를 라우트
@app.route('/predict', methods=['POST'])
def predict():
    input_text = request.form['input_text']
    result =  predict_emtion(input_text)
    return json.dumps(result, ensure_ascii=False)
    # 한글 형태로 유지해주는 코드

# 웹 구동
if __name__ == "__main__":
    app.run(port = 5000, host = "192.168.21.152")
```
