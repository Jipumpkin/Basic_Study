머신러닝 (기계학습)
이전에 있는 데이터를 학습하여 규칙 패턴을 찾아 새로운 데이터를 예측하는 과정

- 머신러닝 종류 3가지
- 지도학습
    정답이 있는 데이터를 학습,예측
    분류: 정답데이터가 범주형(카테고리) -> 클래스 -> 클래스2개 (이진분류), 클래스 3개이상(다중분류)
    회귀: 정답데이터가 연속형(수치형태)
- 비지도학습
    정답이 없는 데이터를 학습, 예측
    군집화, 차원축소
- 강화학습
    보상(점수)을 주는 방향으로 학습


- 머신러닝 7과정
1. 문제정의
2. 데이터수집
3. 데이터 전처리 (이상치,결측치 처리, 특성공학, 단위변환 등)
4. EDA (탐색적 데이터 분석).  데이터 시각화, 기술통계량 등등 . .
5. 모델 선택 및 하이퍼 파라미터 조정
6. 모델 학습
7. 모델 평가
